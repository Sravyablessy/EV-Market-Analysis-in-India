```python
ev_cat= pd.read_csv("/kaggle/input/ev-data/ev_cat_01-24.csv")
print(ev_cat)
print(" ")
print(" Finding is there any Null values")
ev_cat.isnull().sum()
```

```python
# here we remove the first row in the data because it contains all 0's which is unnecessary for us.

ev_cat_new= ev_cat.drop(ev_cat.index[0])
ev_cat_new
```

```python
ev_cat_new.describe()
```

```python
print(ev_cat_new.dtypes)
print(ev_cat_new.shape)
```

```python
# extracting Year from the Date Column

ev_cat_new['Year'] = pd.DatetimeIndex(ev_cat_new['Date']).year
ev_cat_new['Year']
```

```python
ev_cat_new
```

```python
# grouping by year we can find the sum of the EVs category

ev_cat_new_sum = ev_cat_new.groupby('Year').sum(numeric_only=True).reset_index()
ev_cat_new_sum
```

-- maximum in all categories EVs are increased from 2001 to 2023 

```python
# sum of total vehicles over the years based on EV category

cat_sum=ev_cat_new_sum.drop('Year',axis=1).sum()
cat_sum.sort_values(ascending=False)
```

-- Form the above analysis we can say that Two wheeler(NT) has the most demand EV with 2308887 vehicles raised from 2001 to 2024 
-- second and third highest demand Electric vehicles are Three wheeler(T)- 1932683, LIGHT MOTOR VEHICLE- 180330
-- These EVs are mostly adopted/manufactured from 2001 to 2024

-- the least EV category is medium goods vehicle with number of 31 vehicles only from 2001 to 2024

```python
yearly_sum= ev_cat_new_sum.drop('Year',axis=1).sum(axis=1)
yearly_sum
```

```python
# EV Adoption Trends:
# Analyze the adoption rate of electric vehicles over the years.
# How has the demand for EVs changed over time?

yearly_sum_with_year = pd.DataFrame({'Year': ev_cat_new_sum['Year'], 'Yearly Sum': yearly_sum})
yearly_sum_with_year 
```

* overall the EV adoption/manufacturing is relatively incread from 2001 to 2024
* according to the above analysis in 2023 EVs are adopted/manufactured mostly.

```python

import matplotlib.pyplot as plt



# Plotting
plt.figure(figsize=(10,6))
plt.plot(yearly_sum_with_year['Year'], yearly_sum_with_year['Yearly Sum'], marker='o')
plt.title('EVs manufactured/adoption rate over the years')
plt.xlabel('Year')
plt.ylabel('Yearly Sum')
plt.grid(True)

plt.show()
```

```python
ev_sales= pd.read_csv("/kaggle/input/ev-data/ev_sales_by_makers_and_cat_15-24.csv")
print(ev_sales)
print(" ")
print(" Finding is there any Null values")
ev_sales.isnull().sum()
```

```python
print(ev_sales.dtypes)
print(ev_sales.shape)
```

```python
# finding total sales group by Maker(company)

ev_sales_sum = ev_sales.groupby([ 'Maker']).sum(numeric_only=True)
ev_sales_sum
```

```python
# Manufacturer Performance:
# Compare the performance of different EV manufacturers. 
# Which companies have the highest market share, 
# and what factors contribute to their success?

ev_sales_sum['Total Sales(15-24)']= ev_sales_sum.sum(axis=1)
ev_sales_sum.sort_values(by='Total Sales(15-24)',ascending=False)
```

* According to the above analysis we can say that **588266** no.of sales are done by **OLA ELECTRIC TECHNOLOGIES PVT LTD** which has the top performing company in EV market share among all other companies and Total sales done by this company from **2015 to 2024** is the key factor to contribute to its success.
